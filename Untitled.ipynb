```python
# Load the dataset
from data import load_data_viz_data
import pandas as pd
import matplotlib.pyplot as plt

data = load_data_viz_data()
data[['LAT','LONG']] = data[['LAT','LONG']].astype('float64')

print("Shape of the DataFrame:", data.shape)

data.head(10)
```

```python
df_most_common = pd.DataFrame([],columns = ['DISTRICT','MOST_COMMON_INCIDENT'])
i=0
for dis in data['NAME'].unique():
    df = data[data['NAME'] == dis]
    rank = 0 #most common
    off = df[['NAME','OFFENSE_CODE_GROUP']].value_counts().index[rank][1]
    df_most_common.loc[i] = [dis,off]
    i=i+1
df_most_common
```

```python
df_most_common.plot(kind="bar")
```

```python
plt.rcParams["figure.figsize"] = (10, 5)
df_most_common.plot(kind="bar")
plt.xticks(df_most_common['MOST_COMMON_INCIDENT'])
```

```python
df_second_common = pd.DataFrame([],columns = ['DISTRICT','SECOND_COMMON_INCIDENT'])
i=0
for dis in data['NAME'].unique():
    df = data[data['NAME'] == dis]
    rank = 1 #most common
    off = df[['NAME','OFFENSE_CODE_GROUP']].value_counts().index[rank][1]
    df_second_common.loc[i] = [dis,off]
    i=i+1
df_second_common
```

```python
df_second_common['SECOND_COMMON_INCIDENT'].replace('Larceny and vandalism',1, inplace=True) #Replace NaN by "NoAlley"
df_second_common['SECOND_COMMON_INCIDENT'].replace('Police investigation procedure',2, inplace=True) #Replace NaN by "NoAlley"
```
